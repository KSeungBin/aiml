# Decision Tree로 타이타닉 생존자 예측 모델 만들기

```python
# 결측치 채우기 - numerical은 평균값으로 채우기(의사결정을 통해 다른 함수를 사용하기도)
# numerical value
titanic_df['age'].fillna(titanic_df['age'].mean(), inplace=True)
# categorical value
titanic_df['deck'].fillna(titanic_df['deck'].describe()['top'], inplace=True)
titanic_df['embarked'].fillna(titanic_df['embarked'].describe()['top'], inplace=True)
# deck, embarked는 최댓값으로 채운다.
# inplace = True 입력하지 않으면 변수 생성해서 변수에 수정 내용을 저장해야 함
```

```python
titanic_df
```

```python
# str type을 숫자형으로 변환
# preprocessing 라이브러리 import -> LableEncoder 클래스 -> le라는 객체 변수에 대입
# 데이터들을 fit으로 학습시켜, transform으로 레이블을 인코딩 -> 해당 컬럼에 숫자값으로 저장from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
titanic_df['sex'] = le.fit(titanic_df['sex']).transform(titanic_df['sex'])
titanic_df['adult_male'] = le.fit(titanic_df['adult_male']).transform(titanic_df['adult_male'])
titanic_df['alone'] = le.fit(titanic_df['alone']).transform(titanic_df['alone'])
titanic_df['embarked'] = le.fit(titanic_df['embarked']).transform(titanic_df['embarked'])
titanic_df['deck'] = le.fit(titanic_df['deck']).transform(titanic_df['deck'])
titanic_df['who'] = le.fit(titanic_df['who']).transform(titanic_df['who'])
```

```python
titanic_df
```

```python
# drop duplicated columns
# cols가 하나면 리스트 형태로 줄 필요 없음, 2개 이상일 때 리스트 형태로 입력해야 함
# 한번 수행하기, 두 번 수행하면 NO
drop_cols = ["class", "embark_town", "alive"]
titanic_df = titanic_df.drop(drop_cols, axis=1)
titanic_df
```

## 트레이닝 데이터 준비하기

```python
X = titanic_df.iloc[:,1:]
# 'survived'의 index가 0 - target 값(label)이므로 y로 줄 값이기 때문에 X에 넣으면 안 됨
y = titanic_df['survived']
```

```python
# 80%는 트레이닝 데이터, 20%는 테스트 데이터로 나눕니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=45)
```

## 결정 트리 학습모델 만들기

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier()
# 클래스 선언해 객체 생성자 변수를 준다
# NAMING -> regression은 lr, 결정트리 분류기는 dt(c), 분류기는 clf 
# NAMING 잘하면 변수명만 봐도 결정트리 분류기로 학습모델을 만들었음을 알 수 있다.
dt_clf.fit(X_train, y_train)
y_pred = dt_clf.predict(X_test)
print('예측 정확도: %.2f' % accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
# Predict를 수행하고 classification_report() 결과 출력하기
print(classification_report(y_test, y_pred))
# x_test에서 나온 y hat(예측치)을 y_test와 비교하여 성능 측정 -> y_test = y, y_pred = y hat
```
