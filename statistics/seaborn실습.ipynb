```python
# seaborn 임포트,  한글 글꼴 지정
import seaborn as sns
sns.set()
plt.rc('font', family='NanumBarunGothic')
```

```python
 # 내장 데이터 셋 불러오기(seaborn 홈페이지 -> 'API' -> seaborn.load_dataset -> github 확인: csv빼고 이름만 넣어서 출력하면 됨 예)titanic...)
tips = sns.load_dataset('tips')  # pandas에 내장된 데이터프레임을 불러 올 수 있음 
```

```python
tips.head()    # dataframe의 앞부분만 살펴보기
# 데이터 전처리는 시간 오래 걸리고 힘들기 때문에
# 내가 원하는 데이터들을 크롤링(->gathering -> 전처리) 하기 전에
# 오픈되어 있고 공유된, 내장된 데이터(null값 없음. clean한 data)로 머신 러닝을 학습해 모델 만든다. .
```

```python
# tips 데이터프레임의 기술 통계값
#print(tips.mean())
#print(tips.max())
#print(tips.min())
tips.describe()   # describe 함수로 기술 통계값들 한번에 확인 가능
# numerical columns에 대해서만 기술 통계값 describe
```

```python
tips.info() # null값을 확인하기 위해 info 함수 사용
# 결측치 있으면 머신 러닝 할 수 없음
# 데이터 타입을 한눈에 볼 수 있음
```

```python
# 막대그래프 - seaborn 설치하고 여기부터 seaborn 사용
sns.countplot(data=tips, x='day'); # pandas: df에 plot 함수에 kind & title param 주는 것과 같은 기능
plt.title('요일별 팁을 준 횟수');   # sns. 하고 tab키 누르면 메서드와 속성 확인할 수 있음
#seaborn은 plt와 호환됨
```

```python
# boxplot
sns.boxplot(data=tips, x='day', y='tip')
plt.title('요일별 팁의 Box Plot')
```

```python
sns.boxplot(x='day', y='tip', data=tips, hue = 'time');
plt.title('요일별 팁의 Box Plot')
# 머신러닝할 때, 특징값에 대한 결과 값(tip- 우리의 target은 AI모델이 해당 손님이 팁을 얼마나 줄지 예측하게 하는 것)을 같이 학습시킴
# dataset의 features(day, sex...)에서 예측하고자 하는 값(target: tip)을 알아야 함
```

```python
#scatterplot() 
sns.scatterplot(data=tips, x="total_bill", y='tip');
plt.title("식사 금액과 팁의 관계")
```

```python
# scatter과 bar를 합치고 싶을 때에는 seaborn 의 jointplot 사용
# 우선 data로 데이터를 주고, x축 y축에 표기할 데이터를 각각 쓴다.
sns.jointplot(data=tips, x='total_bill',  y='tip');
plt.suptitle("식사 금액과 팁의 관계", y = 1.02) # y에 대한 옵션을 주지 않으면 제목 위치가 어디에 있는지 확인해보기
```

```python
sns.jointplot(data=tips, x='total_bill',  y='tip', kind='hex');  # 시각화를 최적화할 수 있는 kind를 찾기: 자주 쓰이는 그래프 kind는 기억!
plt.suptitle("식사 금액과 팁의 관계", y = 1.02) 
```

```python
# 회귀선을 자동으로 그려주는 sns의 plot함수
sns.jointplot(data=tips, x='total_bill',  y='tip', kind='reg');
plt.suptitle("식사 금액과 팁의 관계", y = 1.02) # hex그래프는 실무에서 잘 사용하지 않지만 회귀 그래프는 많이 사용
# jointplot의 param을 kind='reg'로 설정 or regression 자체 함수 사용
```

```python
#요일별, 성별 팀의 4분위 범위 확인 및 이상치(outlier) 확인
sns.boxplot(data=tips, x='day', y='tip', hue='sex')
plt.title('요일별, 성별 팁의 Box Plot')
```

```python
#violin plot
sns.violinplot(data=tips, x='day', y='tip', hue='sex' )
plt.title('요일별, 성별 팁의 Violin Plot')
```

```python
#sns.distplot() -> sns.displot()으로 버전업 : matplotlib, pandas와 이름 통일하기 위해
sns.displot(data=tips, x='total_bill')
# sns.displot(tips['total_bill'])
```

```python
#kde 옵션: 밀집도 그래프, rug 옵션: 데이터의 위치를 나타내는 선분 표시
sns.displot(data=tips, x='total_bill', kde=True, rug=True)
# rug : 각각의 값이 어디에 해당되는지 표시해줌
```

```python
#kde 옵션: 밀집도 그래프(count수를 표시), rug 옵션: 데이터의 위치를 나타내는 선분 표시, bin : 분포 구분 갯수
sns.displot(data=tips, x='total_bill', kde=True, rug=True, bins=10)
# bins - x축을 10개 범위로 쪼갠다 
# 값을 어떻게 쪼개느냐에 따라 값의 분포가 달라지므로 통찰이 달라짐
# 데이터를 가장 잘 나타낼 수 있는 bin값 설정하는 것 중요
```

```python
#pairplot은 각각 쌍의 그래프를 보여준다.
sns.pairplot(tips) # 서로 다른 features 간 상관관계를 한번에 보여줌
# total_bill과 tip은 선형적 관계에 있다(수렴적 관계로 패턴을 찾을 수 있음)
# total_bill과 size는 선형적 관계가 아니라 패턴을 찾을 수 없음
```

```python
#pairplot은 각각 쌍의 그래프를 보여준다.
# hue = 'sex'의 상관관계도 볼 수 있음.
sns.pairplot(tips, hue='sex') # hue값의 분포를 color별로 보여줌
```

```python
# map 함수 이용 - 코드는 다른 사람들이 봐도 알 수 있게 직관적, 쉽게 구성해야 함
map_sex={'Male':1,'Female':0}
map_smoker = {'Yes':1, 'No':0}
map_day = {
    'Mon': 1,
    "Tue": 2,
    "Wen": 3,
    "Thur": 4,
    "Fri": 5,
    "Sat": 6,
    "Sun": 7
}
map_time={
    'Lunch': 1,
    'Dinner': 2
}
tips['sex'] = tips['sex'].map(map_sex)
tips['smoker'] = tips['smoker'].map(map_smoker)
tips['day'] = tips['day'].map(map_day)
tips['time'] = tips['time'].map(map_time)
```
